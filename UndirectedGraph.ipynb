```python
import pickle
import glob

files = glob.glob('results/ibs/ibs_dicts')
taxa_dicts = []
classifier = 'family'

for file in files:
    with open(file, 'rb') as f:
        taxa_dict = pickle.load(f)
        taxa_dicts.append(taxa_dict)
        
# calculate thresholds

# find average value of each
avg_d = {}
num_files = len(files)

for file in files:
    with open(file, 'rb') as f:
        taxa_dict = pickle.load(f)
    
    total_cells = 0
    for i in taxa_dict[classifier]:
        total_cells += taxa_dict[classifier][i]

    for i in taxa_dict[classifier]:
        if i not in avg_d:
            avg_d[i] = taxa_dict[classifier][i] / (num_files * total_cells)
        else:
            avg_d[i] += (taxa_dict[classifier][i] / (num_files * total_cells))
        
# overall_dict = {}

total_files_per_sample = {}

for file in files:
    with open(file, 'rb') as f:
        taxa_dict = pickle.load(f)
    for i in taxa_dict[classifier]:
        if taxa_dict[classifier][i] < avg_d[i]:
            continue
        else:
            if i not in total_files_per_sample:
                total_files_per_sample[i] = {}
            for j in taxa_dict[classifier]:
                if taxa_dict[classifier][j] >= avg_d[j]:
                    if j not in total_files_per_sample[i]:
                        total_files_per_sample[i][j] = 1
                    else:
                        total_files_per_sample[i][j] += 1   
    
# graphing
import networkx as nx 
import matplotlib.pyplot as plt

graph = nx.Graph()
used = []
for node in total_files_per_sample:
    if node not in used:
        graph.add_node(node, node_size=0)
    used.append(node)    

for node in total_files_per_sample:
    for other in total_files_per_sample[node]:
        if node != other:
            graph.add_edge(node, other, weight = 0.1, color = 'red')

plt.figure(figsize=(18,18))

pos = nx.circular_layout(graph)
pos_higher = {}

offset = 0.05

for k, v in pos.items():
    if(v[1] > 0 and v[0] > 0):
        pos_higher[k] = (v[0] + offset, v[1] + offset)
    elif (v[1] > 0):
        pos_higher[k] = (v[0] - offset, v[1] + offset)
    elif (v[1] <= 0 and v[0] > 0):
        pos_higher[k] = (v[0] + offset, v[1] - offset)
    else:
        pos_higher[k] = (v[0] - offset, v[1] - offset)

nx.draw_circular(graph, node_size = 1, font_size = 8, 
                 label = ("family for " + file1 + ", " + file2 + ", " + file3), width = 0.5)

nx.draw_networkx_labels(graph, pos_higher)

plt.show()
```
