Read in data from classification files produced by the bash script `classifier_script.sh` and output a list of taxa.

```python
# import statements, setup
import csv
import pickle

filename = "ERR4775526"
to_classify = ["sample", "domain", "phylum", "class", "order", "family", "genus"]
```

```python
# parse RDP result

classified = open("RDPtools_data/init_proc/classified.txt", "r")

file = open('results/ibs/' + filename + '.csv', "w", newline='')
file.truncate(0)
writer = csv.writer(file)
lines = classified.readlines()

writer.writerow(to_classify)

taxa_dict = {}
for item in to_classify:
    taxa_dict[item] = {}

# write classifier to csv, save dictionary
for line in lines:
    split = line.split()
    write = [split[0]]
    for i in range(len(split)):
        if split[i] in to_classify:
            write.append(split[i-1])
            if split[i-1] not in taxa_dict[split[i]]:
                taxa_dict[split[i]][split[i-1]] = 1
            else:
                taxa_dict[split[i]][split[i-1]] += 1
    writer.writerow(write)

classified.close()
file.close()

# save dictionary to file

with open('results/ibs/ibs_dicts/' + filename + '.pkl', 'wb') as f:
    pickle.dump(taxa_dict, f)
```
