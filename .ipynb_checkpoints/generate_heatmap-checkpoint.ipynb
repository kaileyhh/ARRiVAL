```python
import pickle
import glob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

files = glob.glob('results/ibs/ibs_dicts/*')
to_classify = ["sample", "domain", "phylum", "class", "order", "family", "genus"]
classifier = "class"
```

```python
# calculate thresholds

# find average value of each
avg_d = {}
num_files = len(files)

for file in files:
    with open(file, 'rb') as f:
        taxa_dict = pickle.load(f)
    
    total_cells = 0
    for i in taxa_dict[classifier]:
        total_cells += taxa_dict[classifier][i]

    for i in taxa_dict[classifier]:
        if i not in avg_d:
            avg_d[i] = taxa_dict[classifier][i] / (num_files * total_cells)
        else:
            avg_d[i] += (taxa_dict[classifier][i] / (num_files * total_cells))
```

```python
# overall_dict = {}

total_files_per_sample = {}

for file in files:
    with open(file, 'rb') as f:
        taxa_dict = pickle.load(f)
    for i in taxa_dict[classifier]:
        if taxa_dict[classifier][i] < avg_d[i]:
            continue
        else:
            if i not in total_files_per_sample:
                total_files_per_sample[i] = {}
            for j in taxa_dict[classifier]:
                if taxa_dict[classifier][j] >= avg_d[j]:
                    if j not in total_files_per_sample[i]:
                        total_files_per_sample[i][j] = 1
                    else:
                        total_files_per_sample[i][j] += 1
```

```python
df = pd.DataFrame.from_dict(total_files_per_sample, orient='index')

plt.figure(figsize=(15, 15), dpi=600)
heatmap = sns.heatmap(df, annot = False, cmap = sns.color_palette("coolwarm", as_cmap=True))
heatmap.set_title("Correlation between species ("+classifier+")")
# heatmap.tick_params(bottom = False)
```
